# 울산-간절곶

```python
gan.info()
```

```python
start_date = pd.to_datetime('2021-01-01 00:00:00')
end_date = pd.to_datetime('2022-06-30 23:00:00')

time_range = pd.date_range(start=start_date, end=end_date, freq='H')
time = pd.DataFrame({'kma_pagobuoy_train.yyyymmddhhmi': time_range})

time['kma_pagobuoy_train.yyyymmddhhmi'] = time['kma_pagobuoy_train.yyyymmddhhmi'].dt.strftime('%Y%m%d%H')
```

```python
time.info()
```

```python
# gan의 시간 데이터는 int type,time_column의 데이터 타입은 object이므로 int로 변경
time['kma_pagobuoy_train.yyyymmddhhmi'] = time['kma_pagobuoy_train.yyyymmddhhmi'].astype(int)
```

```python
gan_m = pd.merge(time,gan,how='outer',on='kma_pagobuoy_train.yyyymmddhhmi')
```

```python
#outer 조인을 해 gan의 데이터의 시간이 중간에 빠지는 시간 없이 1시간 단위로 맞춰줌
gan_m
```

```python
gan_m[gan_m.isna().any(axis=1)]
```

```python
#stn_name,stn중에 결측값이 있어 채워주기( 해당 gan 자체는 파고부이 데이터중 관측점이 간절곶인 데이터프레임)
gan_m['kma_pagobuoy_train.stn'] = 22483.0
gan_m['kma_pagobuoy_train.stn_name'] = '간절곶'
```

```python
gan_m['kma_pagobuoy_train.max_wh'] = gan_m['kma_pagobuoy_train.max_wh'].interpolate()
gan_m['kma_pagobuoy_train.sig_wh'] = gan_m['kma_pagobuoy_train.sig_wh'].interpolate()
gan_m['kma_pagobuoy_train.mean_wh'] = gan_m['kma_pagobuoy_train.mean_wh'].interpolate()
```

```python
gan_m.isnull().sum()
```
