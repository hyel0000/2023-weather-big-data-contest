```python
import pandas as pd
import numpy as np
```

```python
# 기상청 파고부이(khnp_buoy)
df = pd.read_csv('/content/gdrive/MyDrive/날씨빅콘/khnp_buoy_train.csv', index_col = 0)
df
```

```python
# 울산
df.info()
```

```python
# 수치형 자료 중 (-99,-99.9,-999 결측치 값으로 바꾸기)
df.replace(-99, np.nan, inplace=True)
df.replace(-99.9, np.nan, inplace=True)
df.replace(-999, np.nan, inplace=True)
```

```python
df.info()
```

```python
df
```

```python
start_date = pd.to_datetime('2021-01-01 00:00:00')
end_date = pd.to_datetime('2022-06-30 23:59:00')

time_range = pd.date_range(start=start_date, end=end_date, freq='T')
time = pd.DataFrame({'khnp_buoy_train.yyyymmddhhmi': time_range})

time['khnp_buoy_train.yyyymmddhhmi'] = time['khnp_buoy_train.yyyymmddhhmi'].dt.strftime('%Y%m%d%H%M')
```

```python
time.info()
```

```python
# gan의 시간 데이터는 int type,time_column의 데이터 타입은 object이므로 int로 변경
time['khnp_buoy_train.yyyymmddhhmi'] = time['khnp_buoy_train.yyyymmddhhmi'].astype(int)
```

```python
df1 = pd.merge(time,df,how='outer',on='khnp_buoy_train.yyyymmddhhmi')
df1
```

```python
df1
```

```python
df1 = df1.reset_index(drop=True)
df1.index = df1.index + 1
```

```python
df1['khnp_buoy_train.ws'] = df['khnp_buoy_train.ws'].interpolate(method='linear')
df1['khnp_buoy_train.wd'] = df['khnp_buoy_train.wd'].interpolate(method='linear')
```
