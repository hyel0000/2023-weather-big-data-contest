```python
import pandas as pd
import numpy as np
```

```python
# 해양조사원 파고부이(khoa_buoy)
khoa_buoy = pd.read_csv('/content/gdrive/MyDrive/날씨빅콘/khoa_buoy_train.csv', index_col = 0)
khoa_buoy
```

```python
# 부산
khoa_buoy.info()
```

```python
# 수치형 자료 중 (-99,-99.9,-999 결측치 값으로 바꾸기)
khoa_buoy.replace(-99, np.nan, inplace=True)
khoa_buoy.replace(-99.9, np.nan, inplace=True)
khoa_buoy.replace(-999, np.nan, inplace=True)
```

```python
# 부산
khoa_buoy.info()
```

```python
song= khoa_buoy.loc[khoa_buoy['khoa_buoy_train.stn_name'] == 'SongJeong']
```

```python
song
```

```python
start_date = pd.to_datetime('2021-01-01 00:00:00')
end_date = pd.to_datetime('2022-06-30 23:59:00')

time_range = pd.date_range(start=start_date, end=end_date, freq='T')
time = pd.DataFrame({'khoa_buoy_train.yyyymmddhhmi': time_range})

time['khoa_buoy_train.yyyymmddhhmi'] = time['khoa_buoy_train.yyyymmddhhmi'].dt.strftime('%Y%m%d%H%M')
```

```python
time.info()
```

```python
# gan의 시간 데이터는 int type,time_column의 데이터 타입은 object이므로 int로 변경
time['khoa_buoy_train.yyyymmddhhmi'] = time['khoa_buoy_train.yyyymmddhhmi'].astype(int)
```

```python
df = pd.merge(time,song,how='outer',on='khoa_buoy_train.yyyymmddhhmi')
df
```

```python
df['khoa_buoy_train.stn_name'].fillna('SongJeong', inplace = True)
df['khoa_buoy_train.ws'] = df['khoa_buoy_train.ws'].interpolate(method='linear')
df['khoa_buoy_train.wd'] = df['khoa_buoy_train.wd'].interpolate(method='linear')
```

```python
df
```
