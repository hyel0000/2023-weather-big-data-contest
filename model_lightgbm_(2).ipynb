## 전처리

```python
ulsan = ulsan.drop(columns=['ulsan_drag_train_final.num','ulsan_drag_train_final.time','ulsan_drag_train_final.latitude', 'ulsan_drag_train_final.longitude',
                    'kma_pagobuoy_train.stn_name', 'kma_pagobuoy_train.stn','kma_lightbecon_train.stn',
                    'kma_lightbecon_train.stn_name','khnp_buoy_train.stn_name','kma_pagobuoy_train.mean_wh','kma_pagobuoy_train.max_wh'])
```

```python
ulsan_test = ulsan_test.drop(columns=['ulsan_anch_drag_test.area', 'ulsan_anch_drag_test.time','ulsan_anch_drag_test.num', 'ulsan_anch_drag_test.lat','ulsan_anch_drag_test.lon',
                                      'kma_pagobuoy_test.stn_name', 'kma_pagobuoy_test.stn','kma_lightbecon_test.stn','kma_lightbecon_test.stn_name','khnp_buoy_test.stn_name',
                                      'kma_pagobuoy_test.mean_wh','kma_pagobuoy_test.max_wh'])
```

```python
ulsan.columns = ['ulsan_anch_drag_test.sog', 'ulsan_anch_drag_test.cog',
       'ulsan_anch_drag_test.hdg', 'stn_num', 'target',
       'kma_pagobuoy_test.sig_wh',
       'kma_lightbecon_test.wd', 'kma_lightbecon_test.ws', 'khnp_buoy_test.ws',
       'khnp_buoy_test.wd']
```

```python
feature=['ulsan_anch_drag_test.sog', 'ulsan_anch_drag_test.cog',
       'ulsan_anch_drag_test.hdg', 'stn_num',
       'kma_pagobuoy_test.sig_wh',
       'kma_lightbecon_test.wd', 'kma_lightbecon_test.ws', 'khnp_buoy_test.ws',
       'khnp_buoy_test.wd']

y_feature=['target']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le=le.fit(ulsan['stn_num'])
ulsan['stn_num']=le.transform(ulsan['stn_num'])
ulsan_test['stn_num'] = le.transform(ulsan_test['stn_num'])
```

##모델링

```python
from lightgbm.sklearn import LGBMClassifier
from imblearn.over_sampling import SMOTENC
from imblearn.over_sampling import SMOTE
import optuna
from lightgbm import LGBMRegressor, Dataset
from optuna.samplers import TPESampler
from sklearn.model_selection import StratifiedKFold, KFold
from optuna.samplers import TPESampler
from sklearn.metrics import f1_score
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')


test_x = ulsan_test[feature]

ulsan_preds = []
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)

for f, (train_idx, val_idx) in enumerate(tqdm(kf.split(range(len(ulsan)), y=ulsan['target']))):
    train_x = ulsan[feature].iloc[train_idx]
    valid_x = ulsan[feature].iloc[val_idx]
    y_train = ulsan[y_feature].iloc[train_idx]
    y_valid = ulsan[y_feature].iloc[val_idx]


    #sm = SMOTENC(categorical_features=[3], random_state=seed)
    #train_x, y_train = sm.fit_resample(train_x, y_train)

    sm = SMOTE(sampling_strategy=1)
    train_x, y_train = sm.fit_resample(train_x, y_train)



    preds = []
    train_y = y_train[y_feature]
    valid_y = y_valid[y_feature]

    sampler = TPESampler(seed=seed)

    def objective(trial):

            param = {
                'random_state' : seed,
                'metric'    : 'logloss',
                #'max_depth' : trial.suggest_int('max_depth', 10, 25),
                'max_depth' : -1,
                'learning_rate': trial.suggest_float('learning_rate', 1e-3, 0.1),
                'n_estimators': trial.suggest_int('n_estimators', 100, 2000),
                #'num_leaves'  : trial.suggest_int('num_leaves', 31, 1024),
                'num_leaves' : 31,
                'colsample_bytree' : trial.suggest_float('colsample_bytree', 0.5, 1.0),
                'subsample' :trial.suggest_float('subsample', 0.5, 1.0),
                'reg_alpha' : trial.suggest_float('reg_alpha', 0.0, 5.0),
                'reg_lambda' : trial.suggest_float('reg_lambda', 0.0, 5.0),
                'n_jobs' : -1,
                'boost_from_average' : False
            }

            model = LGBMClassifier(**param)
            model.fit(train_x, y_train, eval_set=[(valid_x, y_valid)], eval_metric = 'logloss', verbose=0)

            y_valid_pred = model.predict(valid_x)
            return f1_score(y_valid, y_valid_pred)

    study = optuna.create_study(direction='maximize', sampler=sampler)
    study.optimize(objective, n_trials=10)

    print('Best trial : score {}, \nparams {}'.format(study.best_trial.value, study.best_trial.params))

    model = LGBMClassifier(**study.best_params, random_state=seed, metric='logloss', n_jobs=-1)
    model.fit(train_x, train_y, eval_set=[(valid_x, valid_y)], eval_metric = 'logloss', verbose=0)

    # 확률값 출력
    #pred = model.predict(test_x)
    pred = model.predict_proba(test_x)[:,1]

    preds.append(pred)
    ulsan_preds.append(preds)
```

```python
from lightgbm.sklearn import LGBMClassifier
import optuna
from lightgbm import LGBMRegressor, Dataset
from optuna.samplers import TPESampler
from sklearn.model_selection import StratifiedKFold, KFold
from imblearn.over_sampling import SMOTE
from optuna.samplers import TPESampler
from sklearn.metrics import f1_score
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')


test_x = busan_test[feature]

busan_preds = []
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)

for f, (train_idx, val_idx) in tqdm(enumerate(kf.split(range(len(busan)), y=busan['target']))):
    train_x = busan[feature].iloc[train_idx]
    valid_x = busan[feature].iloc[val_idx]
    y_train = busan[y_feature].iloc[train_idx]
    y_valid = busan[y_feature].iloc[val_idx]


    #sm = SMOTENC(categorical_features=[3], random_state=seed)
    #train_x, y_train = sm.fit_resample(train_x, y_train)
    sm = SMOTE(sampling_strategy=1)
    train_x, y_train = sm.fit_resample(train_x, y_train)

    #sm = RandomUnderSampler()
    #train_x, y_train = sm.fit_resample(train_x, y_train)

    preds = []
    train_y = y_train[y_feature]
    valid_y = y_valid[y_feature]

    sampler = TPESampler(seed=seed)

    def objective(trial):

            param = {
                'random_state': seed,
                'metric'    : 'logloss',
                #'max_depth' : trial.suggest_int('max_depth', 10, 20),
                'max_depth' : -1,
                'learning_rate': trial.suggest_float('learning_rate', 1e-3, 0.1),
                'n_estimators': trial.suggest_int('n_estimators', 100, 2000),
                #'num_leaves'  : trial.suggest_int('num_leaves', 31, 1024),
                'num_leaves' : 31,
                'colsample_bytree' : trial.suggest_float('colsample_bytree', 0.5, 1.0),
                'subsample' :trial.suggest_float('subsample', 0.5, 1.0),
                'reg_alpha' : trial.suggest_float('reg_alpha', 0.0, 5.0),
                'reg_lambda' : trial.suggest_float('reg_lambda', 0.0, 5.0),
                'n_jobs' : -1,
                'boost_from_average' : False
            }

            model = LGBMClassifier(**param)
            model.fit(train_x, y_train, eval_set=[(valid_x, y_valid)], eval_metric = 'logloss', verbose=0)

            y_valid_pred = model.predict(valid_x)
            return f1_score(y_valid, y_valid_pred)

    study = optuna.create_study(direction='maximize', sampler=sampler)
    study.optimize(objective, n_trials=10)

    print('Best trial : score {}, \nparams {}'.format(study.best_trial.value, study.best_trial.params))

    model = LGBMClassifier(**study.best_params, random_state=seed, metric='logloss', n_jobs=-1)
    model.fit(train_x, train_y, eval_set=[(valid_x, valid_y)], eval_metric = 'logloss', verbose=0)

    #pred = model.predict(test_x)
    pred = model.predict_proba(test_x)[:,1]
    preds.append(pred)
    busan_preds.append(preds)
```
