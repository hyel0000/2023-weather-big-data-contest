```python
import math
#e1의 곡선이 항계선의 곡선에 해당하는지 확인
def distance_between_points(center, point):
    # Calculate the distance between the center and the point using their coordinates
    dx = point.x - center.x
    dy = point.y - center.y
    distance = math.sqrt(dx**2 + dy**2)
    return distance


center_latitude = 35.418344   #E1의 2번은 항계선 명세에 나와있는 점이므로 생략
center_longitude = 129.414547
radius = 5436

center_longitude,center_latitude  = transformer.transform(center_longitude,center_latitude)

center = Point(center_longitude, center_latitude)

x = 129.463694
y = 35.446306
x,y  = transformer.transform(x,y)

point = Point(x,y)


distance = distance_between_points(center, point)

print("E1의 결과: ",distance)

if distance >= radius:
    print("해당 좌표는 반경 {}m인 원을 이루는 곡선 위에 위치합니다.".format(radius))
else:
    print("해당 좌표는 반경 {}m인 원의 내부에 위치합니다.".format(radius))

#e2의 곡선이 항계선의 곡선에 해당하는지 확인
x = 129.473861    #E2의 1번은 E1의 2와 같은 점이므로 생략
y =  35.424944
x,y  = transformer.transform(x,y)


distance = distance_between_points(center, point)

print("E2의 결과: ",distance)
if distance >= radius:
    print("해당 좌표는 반경 {}m인 원을 이루는 곡선 위에 위치합니다.".format(radius))
else:
    print("해당 좌표는 반경 {}m인 원의 내부에 위치합니다.".format(radius))

#e3의 곡선이 항계선의 곡선에 해당하는지 확인
x = 129.457333
y =  35.384028
x,y  = transformer.transform(x,y)


distance = distance_between_points(center, point)

print("E3의 결과: ",distance)

if distance >= radius:
    print("해당 좌표는 반경 {}m인 원을 이루는 곡선 위에 위치합니다.".format(radius))
else:
    print("해당 좌표는 반경 {}m인 원의 내부에 위치합니다.".format(radius))
```
